```python
#packages
import numpy as np
import random as rand
import time

#Variables
steps = 1000000
circleArea = 0
squareArea = 0

#functions
def where(x,y):
  return x**2 + y**2

#start time
start = time.time()

#for loop
for i in range(0,steps+1):
  xpos = rand.uniform(0,1)
  ypos = rand.uniform(0,1)

  location = where(xpos,ypos)

  #count how many points land in the circle
  if location <=1:
    circleArea += 1
  squareArea += 1

  #calculate the value of pi
  pi = 4*circleArea/squareArea

end = time.time()

time = end - start

print("The value of pi for " + str(steps) + " steps is "+ str(pi))
print("This simulation took " + str(time) + " seconds to run")
```
